```python
#importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing the IBM Attrition dataset
from google.colab import files
uploaded=files.upload()

ibm_df = pd.read_csv("/punith/Dataset/IBM.csv")

ibm_df.head(20)
```

```python
# Converting Research and Development category in Department Column to Information Technology
ibm_df['Department'] = ibm_df['Department'].replace('Research & Development', 'Information Technology')
ibm_df.head(20)
```

```python
# Information Technology (IT) Department Analysis
# Filter for Information Technology department and making a new dataframe
IT_df = ibm_df[ibm_df['Department'] == 'Information Technology']
IT_df.head()
```

```python
# Count JobSatisfaction levels
IT_job_satisfaction_counts = IT_df['JobSatisfaction'].value_counts().sort_index()

# Print the counts
print("\nJob Satisfaction Levels in Information Technology:")
print(IT_job_satisfaction_counts)

# Bar plot for Job Satisfaction in Research & Development
IT_job_satisfaction_counts.plot(kind='bar', color='black')
plt.title("Job Satisfaction Levels in Information Technology")
plt.xlabel("Job Satisfaction Level")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Count JobSatisfaction levels
IT_education_field_counts = IT_df['EducationField'].value_counts()

# Print the counts
print("\nEducation Field count in Information Technology:")
print(IT_education_field_counts)

# Bar plot for Job Satisfaction in Information Technology
IT_education_field_counts.plot(kind='bar', color='black')
plt.title("Education Field count in Information Technology")
plt.xlabel("Education Field")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Count Marital Status values
IT_marital_status_counts = IT_df['MaritalStatus'].value_counts()

# Print the counts
print("\nMarital Status Counts Information Technology:")
print(IT_marital_status_counts)

# Bar plot for Marital Status in Information Technology
IT_marital_status_counts.plot(kind='bar', color='black')
plt.title("Marital Status Distribution in Information Technology")
plt.xlabel("Marital Status")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Count values in the NumCompaniesWorked column
IT_num_companies_counts = IT_df['NumCompaniesWorked'].value_counts()

# Print the counts
print("\nNumber of Companies Worked Counts (Information Technology):")
print(IT_num_companies_counts)

# Bar plot for NumCompaniesWorked in Information Technology
IT_num_companies_counts.plot(kind='bar', color='black', alpha=0.8)
plt.title("Number of Companies Worked Distribution in Information Technology", fontsize=14)
plt.xlabel("Number of Companies Worked", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Count Work Life Balance levels
IT_Work_Life_Balance_counts = IT_df['WorkLifeBalance'].value_counts().sort_index()

# Print the counts
print("\n Work Life Balance in Information Technology:")
print(IT_Work_Life_Balance_counts)

# Bar plot for Work Life Balance in Information Technology
IT_Work_Life_Balance_counts.plot(kind='bar', color='black')
plt.title("Work Life Balance in Information Technology")
plt.xlabel("Work Life Balance Level")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# These bins categorize employees into income groups for better understanding of salary distribution.
income_bins = [1000, 5000, 10000, 15000, 20000]  # Defining income intervals
income_labels = ["1000-4999", "5000-9999", "10000-14999", "15000-19999"]

# Step 3: Create an "Income_Group" column
# The pd.cut() function groups employees into the defined income intervals and assigns appropriate labels.
IT_df['Income_Group'] = pd.cut(IT_df['MonthlyIncome'], bins=income_bins, labels=income_labels, right=False)

# Step 4: Create a frequency distribution table
# This table shows the number of employees in each income group within the "Research & Development" department.
IT_income_distribution = IT_df['Income_Group'].value_counts().sort_index()

# Step 5: Print the frequency distribution table
# This provides a textual summary of the distribution for quick reference.
print("\nFrequency Distribution Table for Monthly Income in Information Technology:")
print(IT_income_distribution)

# Step 6: Visualize the frequency distribution using a bar chart
# The bar chart visually represents the distribution of income groups for easier interpretation.
plt.figure(figsize=(10, 6))
IT_income_distribution.plot(kind='bar', color='black', edgecolor='black')
plt.title("Frequency Distribution of Monthly Income in Information Technology", fontsize=14)
plt.xlabel("Income Group (INR)", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Frequency Distribution Table for Age
# Objective: Categorize employees into predefined age groups and analyze the frequency distribution
# of employees across these age groups to identify trends in the workforce's age composition.
# Define age intervals and corresponding labels for categorizing employees into age groups
age_bins = [18, 25, 30, 35, 40, 45, 50, 55, 60]  # Define age intervals
age_labels = ["18-24", "25-29", "30-34", "35-39", "40-44", "45-49", "50-54", "55-60"]
IT_df['Age_Group'] = pd.cut(IT_df['Age'], bins=age_bins, labels=age_labels, right=False)

# Create a frequency distribution table
age_distribution = IT_df['Age_Group'].value_counts().sort_index()
print("\nFrequency Distribution Table for Age:")
print(age_distribution)

# Plot the frequency distribution as a bar chart
plt.figure(figsize=(10, 6))
age_distribution.plot(kind='bar', color='black', edgecolor='black')
plt.title("Frequency Distribution of Age", fontsize=14)
plt.xlabel("Age Group", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Calculate the counts for YearsAtCompany
Experience_counts = IT_df['YearsAtCompany'].value_counts()
print("\nYears At Company Counts:")
print(Experience_counts)
plt.figure(figsize=(14, 8))  # Adjust the width and height as needed
sns.countplot(data=IT_df, x='YearsAtCompany', palette='viridis')
plt.title("Experience Distribution", fontsize=16)  # Increased font size for the title
plt.xlabel("Experience in Years", fontsize=14)  # Increased font size for the x-axis label
plt.ylabel("Count", fontsize=14)  # Increased font size for the y-axis label
plt.xticks(fontsize=8)  # Adjust font size and rotation for x-ticks
plt.yticks(fontsize=12)  # Adjust font size for y-ticks
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Optional: Add gridlines for better readability
plt.show()
```

```python
# Calculate the average for Age, MonthlyIncome, and DistanceFromHome
average_values = IT_df[['Age', 'MonthlyIncome', 'DistanceFromHome']].mean()

# Display the results in a styled table
print("Average Values for IT Department:")
table = average_values.to_frame(name="Average Value").style.set_caption("Average Metrics for IT Employees").format("{:.2f}")
display(table)
```

```python
# Attrition by Work-Life Balance
# Objective: Analyze the attrition rates based on employees' work-life balance levels
# and visualize the distribution to understand its impact on attrition.
work_life_table = IT_df.groupby('WorkLifeBalance')['Attrition'].value_counts().unstack()
print("\nAttrition by Work-Life Balance:")
print(work_life_table)

# Bar plot for Work-Life Balance
work_life_table.plot(kind='bar', color=['skyblue', 'orange'])
plt.title("Attrition by Work-Life Balance in IT")
plt.xlabel("Work-Life Balance")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.legend(title="Attrition", loc="upper left")
plt.show()
```

```python
# 4. Attrition by Environment Satisfaction
# Objective: Examine the relationship between environment satisfaction levels and attrition rates
# to understand how workplace environment impacts employee retention.
env_satisfaction_table = IT_df.groupby('EnvironmentSatisfaction')['Attrition'].value_counts().unstack()
print("\nAttrition by Environment Satisfaction:")
print(env_satisfaction_table)

# Bar plot for Environment Satisfaction
env_satisfaction_table.plot(kind='bar', color=['skyblue', 'orange'])
plt.title("Attrition by Environment Satisfaction")
plt.xlabel("Environment Satisfaction")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.legend(title="Attrition", loc="upper left")
plt.show()
```

```python
# 5. Attrition by Job Satisfaction
# Objective: Analyze the relationship between job satisfaction levels and attrition
# to determine how job satisfaction influences employee retention.
# Creating a table to count attrition ('Yes' and 'No') for each job satisfaction level
job_satisfaction_table = IT_df.groupby('JobSatisfaction')['Attrition'].value_counts().unstack()
print("\nAttrition by Job Satisfaction:")
print(job_satisfaction_table)

# Bar plot for Job Satisfaction
job_satisfaction_table.plot(kind='bar', color=['skyblue', 'orange'])
plt.title("Attrition by Job Satisfaction")
plt.xlabel("Job Satisfaction")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.legend(title="Attrition", loc="upper left")
plt.show()
```

```python
# Create a frequency distribution table
Attrition_age_distribution = IT_df.groupby('Age_Group')['Attrition'].value_counts().sort_index().unstack()
print("\nFrequency Distribution Table for Attrition with respect to Age:")
print(Attrition_age_distribution)

# Plot the frequency distribution as a bar chart
plt.figure(figsize=(10, 6))
Attrition_age_distribution.plot(kind='bar', color= ['skyblue', 'orange'])
plt.title("Frequency Distribution of Attrition with respect to Age", fontsize=14)
plt.xlabel("Age Group", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Attrition by Education Field for IT Department
# Objective: Analyze the attrition rates based on employees' education field
# and visualize the distribution to understand its impact on attrition.
education_field_table = IT_df.groupby('EducationField')['Attrition'].value_counts().unstack()
print("\nAttrition by Education Field:")
print(education_field_table)

# Bar plot for Education Field
education_field_table.plot(kind='bar', color=['skyblue', 'orange'], figsize=(10, 6))
plt.title("Attrition by Education Field in IT", fontsize=14)
plt.xlabel("Education Field", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.legend(title="Attrition", loc="upper right")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Attrition by Education
# Objective: Analyze the attrition rates based on employees' education levels
# and visualize the distribution to understand its impact on attrition.
education_table = IT_df.groupby('Education')['Attrition'].value_counts().unstack()

# Print the result for Education
print("\nAttrition by Education:")
print(education_table)

# Bar plot for Education and Attrition
education_table.plot(kind='bar', color=['skyblue', 'orange'])
plt.title("Attrition by Education Level in IT", fontsize=14)
plt.xlabel("Education Level", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=0)
plt.legend(title="Attrition", loc="upper left")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Select only numerical columns for the correlation matrix
numeric_df = IT_df.select_dtypes(include=['number'])

# Calculate the correlation matrix for the numerical columns
correlation_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, cbar=True)

# Adding title and adjusting layout
plt.title("Correlation Heatmap of Numerical Columns", fontsize=14)
plt.tight_layout()

# Show the plot
plt.show()
```
