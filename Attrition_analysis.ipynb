```python
# Import necessary libraries
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# 3. Hyperparameter Tuning Using GridSearchCV

# Define the parameter grid for Random Forest
param_grid = {
    'n_estimators': [100, 200, 300],  # Number of trees in the forest
    'max_depth': [None, 10, 20, 30],  # Maximum depth of the trees
    'min_samples_split': [2, 5, 10],  # Min samples required to split a node
    'min_samples_leaf': [1, 2, 4],    # Min samples required at a leaf node
    'max_features': ['sqrt', 'log2']  # Number of features to consider for the best split
}

# Initialize the Random Forest model
rf_model = RandomForestClassifier(random_state=42)

# Set up GridSearchCV
grid_search = GridSearchCV(
    estimator=rf_model,
    param_grid=param_grid,
    scoring='accuracy',  # Metric to evaluate models
    cv=5,                # 5-fold cross-validation
    n_jobs=-1            # Use all available processors
)

# Fit GridSearchCV on the SMOTE-balanced training data
grid_search.fit(X_train, y_train)

# Get the best model and parameters
best_rf_model = grid_search.best_estimator_
print("Best Parameters:", grid_search.best_params_)

# Model Evaluation

# Make predictions on the test set
y_pred = best_rf_model.predict(X_test)

# Calculate evaluation metrics
accuracy_rf = accuracy_score(y_test, y_pred)
precision_rf = precision_score(y_test, y_pred, pos_label=1)
cm_rf = confusion_matrix(y_test, y_pred)

# Print evaluation metrics
print(f"Random Forest Accuracy: {accuracy_rf:.4f}")
print(f"Random Forest Precision: {precision_rf:.4f}")
print(f"Confusion Matrix:\n{cm_rf}")

# Visualize the confusion matrix as a heatmap
plt.figure(figsize=(6, 6))
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Reds', xticklabels=['No Attrition', 'Attrition'], yticklabels=['No Attrition', 'Attrition'])
plt.title('Random Forest Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
# MODEL USED:
# Random Forest is an ensemble learning method that constructs multiple decision trees during training and outputs the mode of the classes (classification) or mean prediction (regression) of the individual trees. It reduces overfitting, improves accuracy, and handles large datasets with higher dimensionality well. Each tree is trained on a random subset of features, and the final prediction is made through a majority vote or averaging, depending on the task. It is robust to noise and works well on both regression and classification tasks.
# LightGBM (Light Gradient Boosting Machine) is a gradient boosting framework designed for efficiency and scalability. It is optimized for large datasets and can handle categorical features directly. LightGBM builds trees leaf-wise, as opposed to the traditional level-wise approach, which leads to better accuracy but can be more prone to overfitting.
# XGBoost (Extreme Gradient Boosting) is another popular gradient boosting algorithm that has become widely used in machine learning competitions due to its high performance. Like LightGBM, XGBoost builds models by sequentially adding trees to minimize prediction errors, but it uses a level-wise tree growth strategy. It incorporates regularization to reduce overfitting and is highly optimized for performance, supporting both parallelization and distributed computing. XGBoost is versatile and can be used for both classification and regression tasks, and it works well with structured data.
# Support Vector Machine (SVM) is a supervised learning algorithm used for classification and regression tasks. SVM works by finding a hyperplane that best divides a dataset into two classes with the largest possible margin. The algorithm uses a kernel trick to project data into higher-dimensional spaces when classes are not linearly separable.
```
